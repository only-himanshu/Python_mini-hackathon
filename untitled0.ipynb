```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


url = "https://drive.google.com/uc?id=1x8bIz3qNnW5K8-F6HPS_sl7dwyTYlvsL"

df = pd.read_csv(url)

print("Data Preview:\n", df.head())


print("\nColumns:", df.columns)
print("\nMissing values:\n", df.isnull().sum())



df.fillna(df.mean(numeric_only=True), inplace=True)
df.drop_duplicates(inplace=True)

print("\nCleaned Data:\n", df.head())


print("\nSummary:\n", df.describe())


numeric_cols = df.select_dtypes(include=np.number).columns


if len(numeric_cols) >= 2:
    df["Total"] = df[numeric_cols].sum(axis=1)
    df["Average"] = df[numeric_cols].mean(axis=1)


if "Average" in df.columns:
    print("\nTop 5:\n", df.nlargest(5, "Average"))


df[numeric_cols].hist(figsize=(8,6))
plt.suptitle("Distribution of Numeric Data")
plt.show()


if "Average" in df.columns:
    df.nlargest(5, "Average").plot(
        x=df.columns[0], y="Average", kind="bar", title="Top 5 Entries"
    )
    plt.xticks(rotation=30)
    plt.show()


print("\nInsights:")
print("- Data cleaned by handling missing values and duplicates")
print("- Distribution shows spread of data")
print("- Top entries identified using Average values")
```
